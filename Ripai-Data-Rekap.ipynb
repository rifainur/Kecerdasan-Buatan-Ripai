```python
data_map = {data["tahun"]: data["penjualan"] for data in data_rekap}

data_map
```

```python
years = list(data_map.keys())
years.sort()
years
```

```python
data_graph = {}
for year in years:
    year_list = data_graph.get("years", [])
    sale_list = data_graph.get("sales", [])
    year_list.append(year)
    sale_list.append(data_map[year])
    data_graph["years"] = year_list
    data_graph["sales"] = sale_list
data_graph
```

```python
plt.plot(data_graph["years"], data_graph["sales"],'.--', label="Penjualan")
plt.plot(data_graph["years"], [i * 100 for i in range(len(years))],'--', label="Dummy")

plt.xlabel("Tahun")
plt.ylabel("Dalam jutaan rupiah")
plt.title("Data Rekapitulasi Tahunan")
plt.legend()
plt.show()
```

```python
data_map = {data["tahun"]: data["penjualan"] for data in data_rekap}
years = list(data_map.keys())
years.sort()

data_graph = {
    "years": [],
    "sales": [],
    "income": [],
    "profit": []
}

for entry in data_rekap:
    year = entry["tahun"]
    sales = entry["penjualan"]
    income = sales + entry["pendapatan_lain"]
    
    profit = income - entry["operasional"] - entry["pajak"]  # profit = income - operasional - pajak
    
    data_graph["years"].append(year)
    data_graph["sales"].append(sales)
    data_graph["income"].append(income)
    data_graph["profit"].append(profit)

data_graph
```

```python

plt.plot(data_graph["years"], data_graph["sales"], '.--', label="Penjualan")


plt.plot(data_graph["years"], data_graph["income"], 'o-', label="Income")

plt.plot(data_graph["years"], data_graph["profit"], 's-', label="Profit")

plt.xlabel("Tahun")
plt.ylabel("Dalam jutaan rupiah")
plt.title("Data Rekapitulasi Tahunan")

plt.legend()
plt.show()
```
