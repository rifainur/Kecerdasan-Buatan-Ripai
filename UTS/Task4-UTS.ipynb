```python
import json

def get_employees_with_company_country(company_list, employee_list):
    
    company_countries = {company['name']: company['country'] for company in company_list}
    
    employees_with_country = []
    for employee in employee_list:
        company_name = employee['company']
        country = company_countries.get(company_name)
        
        employees_with_country.append({
            "full_name": employee['full_name'],
            "company": company_name,
            "country": country
        })

    return employees_with_country

result = get_employees_with_company_country(company_detail_list, employee_detail_list)
print(json.dumps(result, indent=4))
```
