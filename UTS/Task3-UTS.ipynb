```python
import json

def group_employees_by_domain(company_list, employee_list):
    domain_employees = {}

    company_domains = {company['name']: company['domain'] for company in company_list}

    for employee in employee_list:
        company_name = employee['company']
        domain = company_domains.get(company_name)

        if domain not in domain_employees:
            domain_employees[domain] = []
            
        domain_employees[domain].append(employee['full_name'])

    return domain_employees


result = group_employees_by_domain(company_detail_list, employee_detail_list)
print(json.dumps(result, indent=4))
```
