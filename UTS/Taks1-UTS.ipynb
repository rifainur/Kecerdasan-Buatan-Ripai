```python
def get_sorted_company_list(company_list):
    sorted_companies = sorted(company_list, key=lambda x: x['name'], reverse=True)
    result = [{'name': company['name']} for company in sorted_companies]
    return result
sorted_company_list = get_sorted_company_list(company_detail_list)
print(json.dumps(sorted_company_list, indent=4, separators=(", ", ": ")))
```
