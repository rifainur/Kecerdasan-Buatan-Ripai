```python
def get_employees_with_company_country(company_detail_list, employee_detail_list):
 
    company_nationality_count = {}

    for employee in employee_detail_list:
        company = employee['company']
        nationality = employee['nationality']

        if company not in company_nationality_count:
            company_nationality_count[company] = {}
        
   
        if nationality not in company_nationality_count[company]:
            company_nationality_count[company][nationality] = 0
        
       
        company_nationality_count[company][nationality] += 1

 
    result = []
    for company in company_nationality_count:
        result.append({
            "company": company,
            "employee_nationality": company_nationality_count[company]
        })

    return result


result = get_employees_with_company_country(company_detail_list, employee_detail_list)
print(json.dumps(result, indent=4))
```
